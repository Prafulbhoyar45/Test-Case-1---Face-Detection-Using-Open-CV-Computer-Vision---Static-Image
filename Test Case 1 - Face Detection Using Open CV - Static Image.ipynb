# Test Case 1 - Face Detection Using Open CV - Static Image

```python
# First import Computer vision
import cv2

# Face cascade that give coordinats to rthe CV to fetch the face coordinats in matrix formate
faceCascade = cv2.CascadeClassifier("New Text Document.xml")

# To see the image we use matplotlib
import matplotlib.pyplot as plt
```

```python
# Read image and convert into gray Image
image = cv2.imread("face1.jpg")

# Covert the color of image in Gray
gr_img = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gr_img)
plt.show()

# Detection co-ordinats for face
faces = faceCascade.detectMultiScale(
    gr_img,
    scaleFactor = 1.1,
    minNeighbors = 5,
    minSize = (30,30))

print(faces)

# For color Image
rgb_img = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

# Crop the face
ims = []
for(x,y,w,h) in faces:
    crop = rgb_img[y:y+h,x:x+w]
    ims.append(crop)
    
    
# Show the crop face  
plt.imshow(ims[0])
```
